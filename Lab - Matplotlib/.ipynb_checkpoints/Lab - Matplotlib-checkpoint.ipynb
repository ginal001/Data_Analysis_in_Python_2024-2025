# Plot of Wiener Process

```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's define parameters of our simulation. We are going to use discretization 0.01 and end time 10000. What's more, we are going to plot few realizations.

```python
#Parameters
SEED = 43
T = 10000
STEP = 0.01
N_SIM = 20
N = int(T/STEP)


np.random.seed(SEED)
t = np.linspace(0, T, N + 1)
dW = np.sqrt(STEP)*np.random.randn(N_SIM, N) # N_SIM ralizations of our process = matrix N_SIM X N of normal distribution 
```

```python
W = np.hstack((np.zeros((N_SIM, 1)), np.cumsum(dW, axis = 1)))
up_log = np.sqrt(2 * t[t>3] * np.log( np.log(t[t>3]) ) )  # defining asymptotic from Law of Iterated Logarithm
down_log = -up_log                                            
```

```python
plt.figure(figsize=(13,6))
for i in range(0, N_SIM):
    plt.plot(t, W[i, :])
    
plt.plot(t[t > 3], up_log, '--')
plt.plot(t[t > 3], down_log, '--')
plt.grid()
plt.title("Wiener process with asymptotics")
plt.show()
```
