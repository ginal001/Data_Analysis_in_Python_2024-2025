```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 1000)
```

```python
plt.plot(x ,np.sin(x))
plt.plot(x , x)
plt.title("Plot of sin(x) and x")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)
fig.suptitle('Vertically stacked subplots')
axs[0, 0].plot(x, x)
axs[0, 1].plot(x, -x)
axs[1,0].plot(x, x)
axs[1,1].plot(x, x)
#plt.show()
print("hello world")
```
